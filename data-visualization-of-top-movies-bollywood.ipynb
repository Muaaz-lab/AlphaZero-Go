```python
# Create dataset and visualizations for the user

import pandas as pd
import matplotlib.pyplot as plt

# Create dataset
data = [
    [2017,"Secret Superstar","Musical/Drama",7.8,977],
    [2017,"Tiger Zinda Hai","Action/Thriller",5.9,565],
    [2017,"Hindi Medium","Comedy/Drama",7.8,334],
    [2018,"Sanju","Biographical/Drama",7.7,586],
    [2018,"Padmaavat","Historical/Drama",7.0,585],
    [2019,"War","Action/Thriller",6.5,475],
    [2019,"Kabir Singh","Romance/Drama",7.1,379],
    [2020,"Tanhaji","Historical/Action",7.5,367],
    [2021,"Shershaah","War/Biographical",8.3,250],
    [2022,"Brahmastra","Fantasy/Action",5.6,431],
    [2022,"Drishyam 2","Crime/Thriller",8.2,345],
    [2023,"Jawan","Action/Thriller",7.0,1148],
    [2023,"Animal","Crime/Drama",6.6,915],
    [2024,"Pushpa 2 (Hindi)","Action/Drama",7.2,836],
    [2024,"Stree 2","Horror/Comedy",7.4,875],
    [2024,"Bhool Bhulaiyaa 3","Horror/Comedy",6.8,422],
    [2025,"Dhurandhar","Action/Thriller",7.6,893],
    [2025,"Chhaava","Historical/Drama",7.9,808],
    [2025,"Saiyaara","Romance/Drama",6.9,570]
]

df = pd.DataFrame(data, columns=["Year","Movie","Genre","IMDb_Rating","BoxOffice_Crore"])

# ----- Visualization 1: Genre Distribution -----
plt.figure()
df["Genre"].value_counts().plot(kind="bar")
plt.xlabel("Genre")
plt.ylabel("Number of Movies")
plt.title("Genre Distribution of Top Bollywood Movies (20172025)")
plt.show()

# ----- Visualization 2: Ratings vs Box Office -----
plt.figure()
plt.scatter(df["IMDb_Rating"], df["BoxOffice_Crore"])
plt.xlabel("IMDb Rating")
plt.ylabel("Box Office Collection ( Crore)")
plt.title("IMDb Rating vs Box Office Performance")
plt.show()

df
```
